```python
import pandas as pd

# Load the dataset
file_path = 'Titanic-Dataset.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#Histogram and Boxplots
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
summary_stats = df.describe()

# Visualizations
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

# Histograms and Boxplots
fig, axes = plt.subplots(len(numeric_cols), 2, figsize=(14, 16))
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col].dropna(), ax=axes[i, 0], kde=True)
    axes[i, 0].set_title(f'Histogram of {col}')

    sns.boxplot(x=df[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()


```

```python

# Pairplot to visualize relationships
sns.pairplot(df[numeric_cols + ['Survived']], hue='Survived', diag_kind='kde')
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
#Patterns in data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
#Survival Rate
survival_rate = df['Survived'].value_counts(normalize=True) * 100
print("Survival Rate (%):\n", survival_rate)

sns.countplot(data=df, x='Survived')
plt.title("Survival Distribution")
plt.xticks([0, 1], ['Died', 'Survived'])
plt.show()
```

```python
#Age Distribution And Missing
print("Missing Age values:", df['Age'].isnull().sum())

sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()

sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
#Fare Analysis
print("Max Fare:", df['Fare'].max())
print("Passengers who paid 0 Fare:\n", df[df['Fare'] == 0])

sns.histplot(df['Fare'], bins=40, kde=True)
plt.title("Fare Distribution")
plt.show()

sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare by Class")
plt.show()
```

```python
#Sibsp and Parch
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.countplot(x='SibSp', data=df, ax=ax[0])
ax[0].set_title("Number of Siblings/Spouses")

sns.countplot(x='Parch', data=df, ax=ax[1])
ax[1].set_title("Number of Parents/Children")

plt.tight_layout()
plt.show()
```

```python
#Correlation Matrix
numeric = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']
corr = df[numeric].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# View outliers in fare
high_fare = df[df['Fare'] > 300]
print("\nPassengers with unusually high fare:\n", high_fare[['Name', 'Fare', 'Pclass']])
```
